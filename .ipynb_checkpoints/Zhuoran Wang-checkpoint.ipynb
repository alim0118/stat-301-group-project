## Visualization

```r
#Filter out three languages
project_filter_lan <- projects_data %>% filter(Language == "JavaScript"|
                                              Language == "Python"|
                                              Language == "TypeScript")
project_filter_lan$Language <- factor(project_filter_lan$Language)
```

```r
interquantile_size <- IQR(project_filter_lan$Size)
lower_size <- quantile(project_filter_lan$Size, 0.25) - 1.5 * interquantile_size
upper_size <- quantile(project_filter_lan$Size, 0.75) + 1.5 * interquantile_size
interquantile_issue <- IQR(project_filter_lan$Size)
lower_issue <- quantile(project_filter_lan$Issues, 0.25) - 1.5 * interquantile_issue
upper_issue <- quantile(project_filter_lan$Issues, 0.75) + 1.5 * interquantile_issue

project_filter_lan <- project_filter_lan %>%
  filter(Size >= lower_size & Size <= upper_size) %>%
  filter(Issues >= lower_issue & Issues <= upper_issue)

#Get rid of outliers
```

```r
#Sample the data by different languange
project_filter_lan <- project_filter_lan%>%
  group_by(Language) %>%
  sample_n(min(1000, n()), replace = FALSE)%>%
  ungroup()
head(project_filter_lan)
names(project_filter_lan)[names(project_filter_lan) == "Has Discussions"] <- "HasD"
project_filter_lan$HasD <- as.factor(ifelse(project_filter_lan$HasD == 'TRUE', 1, 0))
```

```r
ggplot(project_filter_lan, aes(x = Issues, y = Stars, color = Language)) +
  geom_point(alpha = 0.5) + 
  facet_wrap(~ HasD) +  
  scale_x_log10() + 
  labs(x = "Number of Issues",
       y = "Number of Stars",
       color = "Language For Programming") +
  theme_minimal() +
  theme(legend.position = "right")
```

#### Relationship between question and graph:
- The faceted scatterplot gives us a direct comparison of how the "has discussion" impacts the github stars, and the issues on the x-axis provides us a visualization of the association of the dots between it and the stars. The dots are colored for different programming languages, it makes distinguishing languages much easier. The use of log scales for the number of issues and stars can deal with the skewness problem in the data. The graph enables us to visualize the association between the inputs: issue, languages and has discussion and the output: stars.
Interpreatation:
- issues: repositories with a medium number of issues seem to have more stars in both facets. As the issues increase, the varations on stars become larger. It might suggest that the issue is positively correlated with the stars. Language:It seems that most of typescript language is at the bottom in both graphs, which might suggest an association between language and stars. has discussion: for the repositories with a medium number of issues, the repositories that have discussions have a wider range of stars.

- Limitations:
The two plots are all overplotting, there are many dots in the same area, so it is difficult to distinguish the individuals and their patterns. In addition, the graph doesn't consider for the impact of other variables. It merely include the three input without considering the confoundings.

## Method and Plan Method:

Inferencing by Multiple linear Regression How to approach(Plan):
1) Set up the basic general form of Multiple linear Regression using lm()function
2) Consider for the interactive terms under different categorical variables: "Programming_languages" and "has_discussion". Firstly, visualize the potential interactions. Then, compute separate multiple linear Regression line for each categorical variables to see if the relation between the input and output changes at different levels of categorical variables. In this way, the interactive terms should be included if the p-value is below the significance level.
3) Identify confounding: Construct and visualize a correlation graph based on all variables in the dataset to determine if any are highly correlated variables with inputs. Add any of the confounding as the additional inputs to a new multiple linear regression and compare it to the original multiple linear regression model. Under the hypothesis, if the p-value is significant, the confounding variable will be included.
4) Consider for the multicollinearity: Create a correlation graph or heat map based on all inputs and calculate the Variance Inflation Factor to evaluate the level of multicollinearity. If there are highly correlated inputs, consider to remove one of them from the inputs. Explanation: Why this model is appropriate: The question is aiming to make inferences. There are three inputs and one output, which fits in the condition for MLR. It is an efficient way to quantify the relationships. The coefficients provide a clear indication of the strength of the association between the input and the output. MLR can accommodate different types of input variables(numerical and categorical variables) and transform them in to either interactive terms or additive terms. In addition, MLR gives a chance to consider for the confounding and multicollinearity in the observational study. Unlike experimental studies, it is impossible to do the random sampling. To make a good fit for the original dataset, MLR allows us to explore the interaction effects as well.

#### Assumptions:
1) the relation between the response and the input variables is linear
2) error terms are independent
3) error terms have a common variance
4) error terms are Normally distributed
5) None of the inputs are correlated In general: Linearity, Independence, Homoscedasticity, Normality and Absence of Multicollinearity. Potential Limits and weakness:
6) MLR can only show the inputs and outputs are associated, but it can't establish a cause-and-effect relationship since it doesn't have a random experiment design to address the causation.
7) MLR is affected by outliers, so a few outliers might draw an incorrect inference.
8) If the model assumptions are not met, the MLR model will not capture the correct relationship.
9) MLR is effected by the number of inputs. If there are too many inputs, it might result in a poor generalization and overfitting. However, if there aren't many inputs, the model might be too simple and underfitting.
10) the accuracy of MLR depends on the data. If the data has large variance, missing data, it can lead MLR to a wrong way.

## Implementation of a proposed model

```r
#basic model
model_additive <- lm(Stars ~ Language + Issues + HasD, data = project_filter_lan)
tidy(model_additive)
```

```r
# Consider for correlation
corr_matrix <- project_filter_lan %>%  

 select_if(is.numeric) %>%
  cor() %>%
  as.data.frame() %>%
  rownames_to_column("var1") %>%
  pivot_longer(-var1, names_to = "var2", values_to = "corr")


plot_corr_matrix <- corr_matrix %>%
  ggplot(aes(var1, var2)) +
  geom_tile(aes(fill = corr), color = "white") +
  scale_fill_distiller("Correlation Coefficient \n",
    palette = "YlOrRd",
    direction = 1, limits = c(-1, 1)
  ) +
  labs(x = "", y = "") +
  theme_minimal() +
  theme(
    axis.text.x = element_text(
      angle = 45, vjust = 1,
      size = 18, hjust = 1
    ),
    axis.text.y = element_text(
      vjust = 1,
      size = 18, hjust = 1
    ),
    legend.title = element_text(size = 18, face = "bold"),
    legend.text = element_text(size = 18),
    legend.key.size = unit(2, "cm")
  ) +
  coord_fixed() +
  geom_text(aes(var2, var1, label = round(corr, 2)), color = "black", size = 6)
plot_corr_matrix
```

```r
# Compare two models
model_considercor <- lm(Stars ~ Language + Issues + HasD + Forks, data = project_filter_lan)
anova(model_additive, model_considercor)
```

```r
# Consider for Multicollinearity
VIF_MLR <- vif(model_considercor)
round(VIF_MLR, 3)
```
